```
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import string

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
```

```
# links to scrape
links = ['https://thegreatestbooks.org/?page=1', 'https://thegreatestbooks.org/?page=2']

# for loop to scrape each link
books = []
for link in links:
  req = requests.get(link)
  soup = BeautifulSoup(req.text, 'lxml')
  scrapes = soup.select('.col h4 a')
  b = [scrape.text for scrape in scrapes]
  books.append(b)
  
# turn list of lists into single list
b_list = [item for sublist in books for item in sublist]

# reshape list with array -- 2 columns in original table
books_authors = np.array(b_list).reshape(int(len(b_list)/2), 2)

# put book and author into dataframe
df = pd.DataFrame(books_authors, columns=['Book', 'Author'])

# get descriptions from links
descriptions = []
for link in links:
  req = requests.get(link)
  soup = BeautifulSoup(req.text, 'lxml')
  scrapes = soup.select('.pb-3 p')
  d = [scrape.text for scrape in scrapes]
  descriptions.append(d)
  
# turn list of lists into single list
d_list = [item for sublist in descriptions for item in sublist]
  
# add description column
df['Description'] = d_list
```
